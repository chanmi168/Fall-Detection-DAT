```
impact_inputdir = '/content/drive/My Drive/中研院/Data/UMAFall/data_management/'+df['DataName'].iloc[50]
df_imp = pd.read_csv(impact_inputdir, header=None)

data_all = np.zeros((251,3,0))
labels_all = np.zeros((0,))
for filename in df['DataName']:
  sub_id = int(filename.split('_')[0])
  activity_id = act_embeddings[filename.split('_')[1]]
  # trial_num = int(filename.split('_')[2])
  location = filename.split('_')[3][:-4]
  # get waist data only
  if location == 'waist':
    impact_inputdir = '/content/drive/My Drive/中研院/Data/UMAFall/data_management/'+filename

    # get only the first 250 samples for now
    df_imp = pd.read_csv(impact_inputdir, header=None).loc[0:250, :]
    if df_imp.shape[0] == 251:
      sub_id = filename.split('_')[0]

      # print(df_imp.shape, filename)
      data_all = np.concatenate((data_all, np.expand_dims(df_imp.to_numpy(), axis=2)), axis=2)
      labels_all = np.append(labels_all, activity_id)
```

```
i = 100
plt.plot(data_all[:,0,i], label='x')
plt.plot(data_all[:,1,i], label='y')
plt.plot(data_all[:,2,i], label='z')
plt.ylabel('some numbers')
plt.xlabel('time')
plt.legend(loc='upper right')
plt.show()
```
